```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col_names = ["sepallength", "sepalwidth", "petallength", "petalWidth", "class"]
ulr = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
dataset = pd.read_csv(ulr, names=col_names)
dataset
```

```python
dataset_knn = dataset.values[:,0:4]
labels = dataset.values[:,-1]
dataset_knn
```

```python
import math

def distance(x1,x2):
  distances = 0.0
  for i in range(len(x1)):
    distances += (x1[i] - x2[i])**2

  return (math.sqrt(distances))
```

```python
testing_data = dataset.values[101,0:4]
```

```python
distance_cumm = []
for i in range(len(dataset_knn)):
  distance_cumm.append([distance(dataset_knn[i], testing_data), labels[i]])
```

```python
distance_cumm.sort()
distance_cumm
```

```python
k = 7
label_topk = []
for i in range(1,k+1):
  label_topk.append(distance_cumm[i][1])

label_topk
```

```python
from collections import Counter
count = Counter(label_topk)
count
```

```python
print("Our final predicted label is", count.most_common(1)[0][0])
```
