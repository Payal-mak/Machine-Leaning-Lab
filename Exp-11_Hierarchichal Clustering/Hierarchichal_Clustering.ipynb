```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset  = pd.read_csv('/content/KMeans.csv')
dataset
```

```python
x = dataset.iloc[:,[1,2]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, linkage='ward')
yhc = hc.fit_predict(x)
yhc
```

```python
# Plot the clusters
plt.scatter(x[yhc == 0,0], x[yhc == 0,1], c='red',    label='Cluster 1')
plt.scatter(x[yhc == 1,0], x[yhc == 1,1], c='yellow', label='Cluster 2')
plt.scatter(x[yhc == 2,0], x[yhc == 2,1], c='blue',   label='Cluster 3')
plt.scatter(x[yhc == 3,0], x[yhc == 3,1], c='green',  label='Cluster 4')
plt.scatter(x[yhc == 4,0], x[yhc == 4,1], c='magenta',label='Cluster 5')
plt.legend()
```
