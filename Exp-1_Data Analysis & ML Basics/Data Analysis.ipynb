```python
data=pd.read_csv("/content/survey_results_public.csv")
data.head()
```

```python
data.shape
```

```python
#Count the number of null values in each feature
#Count the percentage of null values
#Draw the pie chart for the number of people who finds coding as hobby
#Determine the number of people contributing to the open source projects
#Determine the top 20 countries from where the responses are obtained
#Do other 5 analysis as per your own thinking which involves different charts or graphs
data.isnull().sum()/data.shape[0]*100
data.isnull().sum()
```

```python
#Count the number of null values in each feature
null_counts = data.isnull().sum()
print("Null values in each column:\n", null_counts)
```

```python
#Count the percentage of null values
null_percent = (null_counts / data.shape[0]) * 100
print("\nPercentage of null values:\n", null_percent)
```

```python
# 3. Pie chart for number of people who find coding a hobby
hobby_counts = data['Hobby'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(hobby_counts, labels=hobby_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
plt.title('Do you code as a hobby?')
plt.show()
```

```python
# 4. Number of people contributing to open source projects
oss_counts = data['OpenSource'].value_counts()
print("\nOpen source contribution:\n", oss_counts)
oss_counts.plot(kind='bar', color='skyblue')
plt.title('Contribution to Open Source Projects')
plt.xlabel('Response')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.show()
```

```python
# 5. Top 20 countries from where the responses are obtained
top_countries = data['Country'].value_counts().head(20)
top_countries.plot(kind='barh', figsize=(10,8), color='coral')
plt.title('Top 20 Respondent Countries')
plt.xlabel('Number of Responses')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 6. Most popular programming languages
# Split semi-colon separated values and count
langs = data['LanguageWorkedWith'].dropna().str.split(';')
all_langs = pd.Series([lang for sublist in langs for lang in sublist])
top_langs = all_langs.value_counts().head(10)
top_langs.plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Programming Languages')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Career Satisfaction - Count each unique satisfaction level
career_satisfaction_counts = data['CareerSatisfaction'].value_counts()

# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=career_satisfaction_counts.values, y=career_satisfaction_counts.index, palette='viridis')
plt.title('Career Satisfaction Levels')
plt.xlabel('Number of Responses')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
# 8. Distribution of compensation
comp = data['ConvertedSalary'].dropna()
comp = comp[comp < comp.quantile(0.95)]  # Remove top 5% for better visibility
plt.figure(figsize=(10,5))
sns.histplot(comp, bins=50, color='goldenrod')
plt.title('Annual Compensation (Filtered)')
plt.xlabel('Compensation (USD)')
plt.show()
```

```python
# 9. Education level distribution
edu_counts = data['FormalEducation'].value_counts()
edu_counts.plot(kind='barh', figsize=(10,6), color='slateblue')
plt.title('Formal Education Level of Respondents')
plt.xlabel('Number of People')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 10. Most common developer types
dev_types = data['DevType'].dropna().str.split(';')
all_dev_types = pd.Series([role for sublist in dev_types for role in sublist])
top_dev_types = all_dev_types.value_counts().head(10)
top_dev_types.plot(kind='bar', color='darkcyan')
plt.title('Top 10 Developer Types')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.show()
```
